$\textbf{Graphs}$
---
- A graph $G$ consists of:
  - A nonempty set $V(G)$ of vertices.
  - A set $E(G)$ of edges.

- Each edge is associated with a set of one or two vertices, referred to as its endpoints. This relationship is defined by the edge-endpoint function.

**Key points:**

- An edge with a single endpoint is called a **loop**.
- Multiple edges with the same endpoints are termed **parallel** edges.
- An edge connects its endpoints, and vertices connected by an edge are called **adjacent**.
- A vertex that is an endpoint of a loop is adjacent to itself.
- An edge is **incident** on each of its endpoints.
- Two edges incident on the same endpoint are **adjacent**.
- A vertex with no incident edges is termed **isolated**.

\
$EXAMPLE$

Consider the following graph:


```python
%matplotlib inline

import matplotlib.pyplot as plt
import networkx as nx

# Create a simple graph
G = nx.Graph()

# Add nodes to the graph
nodes = [1, 2, 3, 4, 5, 6]
G.add_nodes_from(nodes)

# Define edges between nodes
edges = [(1, 2), (1, 3), (2, 3), (2, 4), (3, 4), (4, 5), (5,6)]
G.add_edges_from(edges)

# Compute the layout for the graph
pos = nx.spring_layout(G)  # spring_layout is generally fast and effective

# Draw the graph
nx.draw(G, pos, with_labels=True, node_color='pink', node_size=800, font_size=12, font_color='brown', edge_color='green')

# Display the graph
plt.show()
```

In graph theory, a graph ( $ G $ ) is a collection of vertices (or nodes) connected by edges. Shown above is a simple graph with six vertices and seven edges. Here’s how the graph concepts are illustrated:

- **Vertices (Nodes)**: Represented by the red circles, each labeled with a unique number (1 to 6).
- **Edges**: Represented by the gray lines connecting the nodes. Each edge connects two vertices, illustrating adjacency.
- **Adjacency**: Nodes connected by an edge are adjacent. For example, node 1 is adjacent to nodes 2 and 3.
- **Incidence**: Each edge is incident on its endpoints. For example, the edge \( (1, 2) \) is incident on nodes 1 and 2.
- **Isolated Vertices**: This graph has no isolated vertices; every node has at least one edge.

By visualizing this graph, you can observe the relationships between vertices and edges, which are foundational concepts in graph theory.

$\textbf{Directed Graph (Digraphs)}$
----
In a directed graph, or digraph:

- There is a nonempty set $(V(G))$ of vertices.
- There is a set $(D(G))$ of directed edges.
- Each directed edge is associated with an ordered pair of vertices.
- If edge $e$ corresponds to the pair $(v, w)$ of vertices, $e$  is designated as the directed edge from $v$ to $w$.

\
$EXAMPLE$

Consider the following graph:



```python
# Ensure inline plotting in Jupyter Notebook
%matplotlib inline

import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
DG = nx.DiGraph()

# Add nodes to the graph
nodes = [1, 2, 3, 4, 5]
DG.add_nodes_from(nodes)

# Define directed edges between nodes
edges = [(1, 2), (1, 3), (3, 2), (2, 4), (3, 4), (4, 5)]
DG.add_edges_from(edges)

# Compute the layout for the graph
pos = nx.spring_layout(DG)  # spring_layout is generally fast and effective

# Draw the directed graph
nx.draw(DG, pos, with_labels=True, node_color='pink', node_size=800, font_size=12, font_color='brown', edge_color='green', arrowsize=20, arrowstyle='-|>')

# Display the graph
plt.show()
```

The directed graph illustrated by the code consists of the following characteristics:

- **Vertices (Nodes)**: There are five vertices represented by numbered nodes (1, 2, 3, 4, 5).

- **Directed Edges**: The edges between the vertices are directed, indicating a one-way relationship between them. For instance, there's an edge from node 1 to node 2, from node 1 to node 3, and so forth. Each directed edge is denoted by an arrow indicating the direction of the relationship.

- **Edge Directions**: For instance, the edge from node 1 to node 2 means there is a directed relationship from node 1 to node 2. This implies a directional flow of influence or connectivity between the nodes.

- **Connectivity**: The directed edges establish connectivity between the nodes, defining a directional flow or relationship among them.

- **Graph Layout**: The layout of the graph is computed using the spring layout algorithm, resulting in an arrangement of nodes that minimizes overlapping edges while preserving edge lengths as much as possible.

- **Visualization**: Nodes are depicted as pink circles labeled with their corresponding numbers. Directed edges are drawn as green arrows, indicating the direction of the relationship. The arrowheads point from the source node (where the edge starts) to the target node (where the edge ends).

- **Arrows**: The arrows on the edges indicate the direction of the relationship. For instance, an arrow pointing from node 1 to node 2 signifies that there is a directed edge from node 1 to node 2.

This directed graph visually represents a network of nodes connected by directed edges, where the direction of each edge indicates a specific relationship or flow between the associated vertices.

----
Let $n$ be a positive integer. A complete graph on $n$ vertices, denoted $K_n$, is a simple
graph with $n$ vertices and exactly one edge connecting each pair of distinct vertices.

$\textbf{Bipartite Graphs}$
----
Let $n$ be a positive integer. A complete graph on $n$ vertices, denoted $K_n$, is a simple
graph with $n$ vertices and exactly one edge connecting each pair of distinct vertices.

Let $m$ and $n$ be positive integers. A complete bipartite graph on $(m, n)$ vertices, denoted $K_{m,n}$, is a simple graph with distinct vertices $v_1, v_2, . . . , v_m$ and $w_1, w_2, . . . , w_n$ that satisfies the following properties: $\forall i,k = 1,2,...,m$ and $\forall j,l = 1,2,...,n$

1. There is an edge from each vertex $v_i$ to each vertex $w_j$.
2. There is no edge from any vertex $v_i$ to any other vertex $v_k$.
3. There is no edge from any vertex $w_j$ to any other vertex $w_l$.

\
$EXAMPLE$

Consider the following graph:

```python
import matplotlib.pyplot as plt
import networkx as nx

kabit_network = nx.Graph()
kabit_network.add_nodes_from([1,3,5], bipartite=0) # Add the node attribute "bipartite"
kabit_network.add_nodes_from([2,4], bipartite=1)
kabit_network.add_edges_from([(1,2), (1,4)
                              , (3,2), (3,4)
                              , (5,2), (5,4)])

top = nx.bipartite.sets(kabit_network)[0]
pos = nx.bipartite_layout(kabit_network, top)
nx.draw(kabit_network, pos=pos, with_labels=True, node_color=['pink','pink','pink','pink','pink'])
plt.show()
```

The directed graph illustrated here represents a relationship network, specifically a bipartite graph. In this graph:

- There are two distinct sets of nodes: one set consists of nodes 1, 3, and 5, while the other set comprises nodes 2 and 4. These sets are indicative of a bipartite structure, where nodes from one set are only connected to nodes from the other set, but not within the same set.
- Nodes 1, 3, and 5 are connected to nodes 2 and 4, forming directed edges from nodes in the first set to nodes in the second set.
- The edges signify relationships or connections between entities represented by the nodes. For example, if node 1 represents a person and node 2 represents a place, an edge from node 1 to node 2 could represent a person visiting that place.
- The graph layout is organized in such a way that nodes from both sets are visually distinct, aiding in the interpretation of the bipartite structure.
The visualization depicts the relationships between the nodes, allowing for the analysis of connectivity patterns within the network.

$\textbf{Subgraph}$
----
- A graph $H$ is set to be the subgraph of graph $G$ when every vertices and edges in $H$ are also in $G$, and every edge in $H$ also follows the same endpoints in $G$.

 The first graph(Graph $F$) below is a complete graph of 4 vertices $K_4$. The second graph(Graph $S$) is a subgraph of the graph $F$ since all vertices in graph $S$ is also in graph $F$, and since graph $F$ is complete graph, all endpoint of edge in graph $S$ is also in graph $F$.

\
$EXAMPLE$

Consider the following graph:

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.Graph()


G.add_nodes_from([1,2,3,4])
edgelist = [(1,2),(1,3),(1,4),(2,3),(2,4),(3,4)]
new = [(1,2),(2,3),(3,4)]
plt.figure(figsize=(20, 5))

plt.subplot(121)
nx.draw(G, pos=nx.circular_layout(G), with_labels=True)
nx.draw_networkx_edges(G, pos=nx.circular_layout(G), edgelist=edgelist)
plt.title('Original Graph')

plt.subplot(122)
nx.draw(G, pos=nx.circular_layout(G), with_labels=True)
nx.draw_networkx_edges(G, pos=nx.circular_layout(G), edgelist=new, edge_color='red')
plt.title('Subgraph')

plt.show()
```

The directed graph depicted by the code consists of four vertices labeled 1, 2, 3, and 4. The edges are represented by pairs of vertices, indicating the direction from one vertex to another.

In the original graph, all vertices are interconnected, forming a complete graph. Specifically, vertex 1 is connected to vertices 2, 3, and 4. Vertex 2 is connected to vertices 1, 3, and 4. Vertex 3 is connected to vertices 1, 2, and 4. Vertex 4 is connected to vertices 1, 2, and 3. This configuration demonstrates a symmetrical relationship among all vertices.

In the subgraph, a subset of edges is highlighted in red. This subgraph comprises three directed edges: (1, 2), (2, 3), and (3, 4). These edges form a directed path that traverses through vertices 1, 2, 3, and 4 sequentially. This subgraph illustrates a linear sequence of directed edges, showcasing a directed path within the original graph.

$\textbf{Travelling in graph}$
----
Let example graph be Graph $G$ and let $v$ and $w$ be vertices in a graph $G$.
 - A Walk in a graph is a finite alternative sequence of moving to the incident edge and vertex from $v$ to $w$.
 - A Trail is a walk with no repeated edges.
 - A Path is a walk with no repeated vertex.

 - A Closed Walk is a walk that starts and end in same vertex.
 - A Circuit is a closed walk with atleast one edge and with no repeated edge.
 - A Simple Circuit is a closed walk with no repeated vertex exceot the start and the end.

\
$EXAMPLE$

Consider the following graph:


```python
import matplotlib.pyplot as plt
import networkx as nx
G.add_nodes_from([1,2,3,4])

walk = [(1,2),(2,3),(3,4),(3,1),(1,4)]
trail = [(1,2),(2,3),(3,4),(4,2)]
path = [(1,2),(2,3),(3,4)]
closed_walk = [(1,2),(2,4),(4,3),(3,1)]
circuit = [(1,2),(2,3),(3,4),(4,1)]
simple_circuit = [(1,2),(2,3),(3,4),(4,1)]

plt.subplot(231)
nx.draw(G, pos=nx.circular_layout(G), with_labels=True)
nx.draw_networkx_edges(G, pos=nx.circular_layout(G), edgelist=walk, edge_color='red', width=2)
plt.title('Walk')

plt.subplot(232)
nx.draw(G, pos=nx.circular_layout(G), with_labels=True)
nx.draw_networkx_edges(G, pos=nx.circular_layout(G), edgelist=trail, edge_color='red', width=2)
plt.title('Trail')

plt.subplot(233)
nx.draw(G, pos=nx.circular_layout(G), with_labels=True)
nx.draw_networkx_edges(G, pos=nx.circular_layout(G), edgelist=path, edge_color='red', width=2)
plt.title('Path')

plt.subplot(234)
nx.draw(G, pos=nx.circular_layout(G), with_labels=True)
nx.draw_networkx_edges(G, pos=nx.circular_layout(G), edgelist=closed_walk, edge_color='red', width=2)
plt.title('Closed Walk')

plt.subplot(235)
nx.draw(G, pos=nx.circular_layout(G), with_labels=True)
nx.draw_networkx_edges(G, pos=nx.circular_layout(G), edgelist=circuit, edge_color='red', width=2)
plt.title('Circuit')

plt.subplot(236)
nx.draw(G, pos=nx.circular_layout(G), with_labels=True)
nx.draw_networkx_edges(G, pos=nx.circular_layout(G), edgelist=simple_circuit, edge_color='red', width=2)
plt.title('Simple Circuit')

plt.tight_layout()
plt.show()
```

$\textbf{Connectedness}$
----
Let example graph be Graph $G$ and let $v$ and $w$ be vertices in a graph $G$.
 - $v$ and $w$ is connected $\iff$ there is a walk from $v$ to $w$.
 - $G$ is a connected graph when any vertices $v$ to $w$ is connected.
 - If $G$ is a connected graph, then any two distinct vertices in $G$ can be a path.
 - If there is circuit in G, then removing one of any edge from the circuit will create a trail.
 - If $G$ is a connected graph and has a curcuit, then it is possible to remove one edge from that curcuit without disconnecting $G$.

\
$\textbf{Connected component}$
---
 - A graph $H$ is a connected component of $G$ $\iff$ $H$ is a subgraph of $G$, $H$ is connected, and no connected subgraphs of $G$ has $H$ as a subgraphs and contains vertices or edges that are not in $H$.

 \
$EXAMPLE$

Consider the following graph:

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.Graph()


G.add_nodes_from([1,2,3])
edgelist = [(1,2),(2,3),(3,1)]
comp = [(1,2),(2,3)]
plt.figure(figsize=(20, 5))

plt.subplot(121)
nx.draw(G, pos=nx.circular_layout(G), with_labels=True)
nx.draw_networkx_edges(G, pos=nx.circular_layout(G), edgelist=edgelist)
plt.title('Original Graph')

plt.subplot(122)
nx.draw(G, pos=nx.circular_layout(G), with_labels=True)
nx.draw_networkx_edges(G, pos=nx.circular_layout(G), edgelist=comp, edge_color='red')
plt.title('Connected Component')

plt.show()
```

$\textbf{Euler's Circuit and Trail}$
----
- Euler's Circuit is a closed walk in a graph that uses every vertex in $G$ atleast once and uses every edge EXACTLY ONCE.
 - In a graph $G$, there is a Euler's Circuit $\iff$ every degree of every vertex in $G$ is positve even number.
 ---
 - Euler's Trail is a walk in a graph that passes through every edge EXACTLY ONCE, and atleast once every vertices.

\
$EXAMPLE$

Consider the following graph:

```python
import matplotlib.pyplot as plt
import networkx as nx
plt.figure(figsize=(20, 5))

G = nx.Graph()
G.add_nodes_from([1,2,3,4,5])
Gnew = nx.Graph()
Gnew.add_nodes_from([1,2,3,4])
edgelist = [(1,2),(2,3),(3,1),(4,5),(1,4),(2,4),(2,5),(3,4),(1,5),(3,5)]
edgelist2 = [(1,2),(2,4),(4,3),(3,2)]

plt.subplot(121)
nx.draw(G, pos=nx.circular_layout(G), with_labels=True)
nx.draw_networkx_edges(G, pos=nx.circular_layout(G), edgelist=edgelist, edge_color='red')
plt.title('Eulers Circuit')

plt.subplot(122)
nx.draw(Gnew, pos=nx.circular_layout(Gnew), with_labels=True)
nx.draw_networkx_edges(Gnew, pos=nx.circular_layout(Gnew), edgelist=edgelist2, edge_color='red')
plt.title('Eulers Trail')

plt.show()
```
